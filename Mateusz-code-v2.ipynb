```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import curve_fit, bisect
from scipy.special import factorial, factorial2
import matplotlib.pyplot as plt
import os
import time
from tqdm import tqdm
```

```python
eV = 1.602176634 * 1e-19
keV = eV * 1e3
MeV = eV * 1e6
GeV = eV * 1e9
TeV = eV * 1e12
h_bar = 6.582119569 * 1e-16 * eV
c = 299792458
m_chi = 1 * TeV
m_phi = 1 * GeV
delta = 100 * keV
alpha = 1/137
v = 130
r_0 = 1 / (alpha * m_chi * c / h_bar)

l = 0
```

```python
# Yukawa potential
def Yukawa(r, eps_phi):
    return np.exp(-eps_phi * r) / r
```

```python
class one_kinematic_point_N2:
    
    def __init__(self, eps_v=0.2, eps_delta=0.1, eps_phi=0.2, potential=Yukawa, num_acc=5,
                 r_min=1e-3, r_max=1e3, n=int(1e5), eps_damping=0.0, update_acc=True):
        
        self.eps_v = eps_v
        self.eps_delta = eps_delta
        self.eps_phi = eps_phi
        self.below_threshold = True if self.eps_v**2 < self.eps_delta**2 + alpha**2/4 * self.eps_delta**4 else False
        self.V = potential
        self.num_acc = num_acc
        self.eps_damping = np.min((self.eps_v, np.sqrt(self.eps_delta**2 - self.eps_v**2)))/100 if self.below_threshold else eps_damping
        self.damping = False if self.eps_damping==0 else True
        self.update_accuracy = update_acc # used to tweak model parameters in case of insufficient accuracy detected
        
        self.r_min = r_min
        self.r_max = r_max / 100 if self.below_threshold else r_max     # r_min, r_max: interval of integration
        self.n = n             # n: resolution of the solution (number of points between r_min, r_max in log scale) 
        self.r = np.geomspace(self.r_min, self.r_max, self.n)
                               # Logarithmic mesh of the domain of the wavefunction
        
        self.ratio_optimal = None # Used when below_threshold
        self.Schr_sol = None   # Linearly independent solutions with regular boundary conditions 
        self.psi = None        # Regular solutions to r*R (I suppose)
        self.delta = None      # phase shifts delta_{n,i}
        self.beta = None       # betas_{n,i}
        self.S = None          # S matrix associated with the given kinematical point
        self.unitarity = None  # measures how far S-matrix is from being unitary; 0.0 means perfect unitarity
        self.cs = None         # Scattering cross-section
        self.Sommerfeld = None # Sommerfeld enhancement factor
                
    def solve_Schr_eq(self):
        
        def rhs_Schr_eq(r, X):
            if self.damping:
                equation = np.array([
                                    [0, 0, 1, 0],
                                    [0, 0, 0, np.exp(-self.eps_damping*r**2)*1],
                                    [l*(l+1)/r**2 - self.eps_v**2, -self.V(r, self.eps_phi), 0, 0],
                                    [np.exp(-self.eps_damping*r**2)*(-self.V(r, self.eps_phi)), np.exp(-self.eps_damping*r**2)*(l*(l+1)/r**2 + self.eps_delta**2 - self.eps_v**2), 0, 0]
                                    ])
            else:
                equation = np.array([
                                    [0, 0, 1, 0],
                                    [0, 0, 0, 1],
                                    [l*(l+1)/r**2 - self.eps_v**2, -self.V(r, self.eps_phi), 0, 0],
                                    [-self.V(r, self.eps_phi), l*(l+1)/r**2 + self.eps_delta**2 - self.eps_v**2, 0, 0]
                                    ])
            return np.dot(equation, X)    
        
        # The derivative of psi_11
        def phi11_dot(r, X):
            return X[2]

        # The derivative of psi_22
        def phi22_dot(r, X):
            return X[3]

        rs = (self.r_min, self.r_max)
        sol1 = solve_ivp(rhs_Schr_eq, rs, np.array([1, 0, (l+1)/self.r_min, 0]), 
                         t_eval = self.r, events = (phi11_dot, phi22_dot), method = 'Radau', rtol = 10**(-self.num_acc))
        sol2 = solve_ivp(rhs_Schr_eq, rs, np.array([0, 1, 0, (l+1)/self.r_min]), 
                         t_eval = self.r, events = (phi11_dot, phi22_dot), method = 'Radau', rtol = 10**(-self.num_acc))
        
        self.Schr_sol = sol1, sol2
        self.psi = sol1.y[0], sol1.y[1], sol2.y[0], sol2.y[1]    
        # psi_11^(1) (11 state, 1st linearly indep. sol.) psi_22^(1) psi_11^(2) psi_22^(2)
        
        if self.below_threshold:
            
            def chi(ratio):
                sol = solve_ivp(rhs_Schr_eq, rs, ratio * np.array([1, 0, (l+1)/self.r_min, 0]) - np.array([0, 1, 0, (l+1)/self.r_min]),
                                t_eval = self.r, events = (phi11_dot, phi22_dot), method = 'Radau', rtol = 10**(-self.num_acc))
                return sol.y[1,-1]
        
            ratio_lower, ratio_upper = np.longdouble(0.9 * self.Schr_sol[1].y[1,-1] / self.Schr_sol[0].y[1,-1]), np.longdouble(1.1 * self.Schr_sol[1].y[1,-1] / self.Schr_sol[0].y[1,-1])
            damping = self.damping
            self.damping = False # so that the good version of rhs is passed to chi()
            self.ratio_optimal = bisect(chi, ratio_lower, ratio_upper)
            self.damping = damping 
            
            self.Schr_sol[0] = solve_ivp(rhs_Schr_eq, rs, y0 = self.ratio_optimal * np.array([1, 0, (l+1)/self.r_min, 0]) - np.array([0, 1, 0, (l+1)/self.r_min]),
                                         t_eval = self.r, events = (phi11_dot, phi22_dot), method = 'Radau', rtol = 10**(-self.num_acc))
               
    def plot_wavefun(self):
        
        titles = ['$\Psi_{11}^{(1)}$','$\Psi_{22}^{(1)}$','$\Psi_{11}^{(2)}$','$\Psi_{22}^{(2)}$']
        fig, ax = plt.subplots(4, sharex='all', sharey='all', figsize=(16,10))
        for idx, pl_title in enumerate(titles):
            ax[idx].plot(self.r, self.psi[idx])
            ax[idx].set_ylabel(pl_title)
            ax[idx].set_xscale('log', base=10)
        save_path = os.path.join(os.getcwd(),'plots',
                                 'eps_v={}_eps_delta={}_eps_phi={}_l={}.png'.format(self.eps_v, self.eps_delta, self.eps_phi, l))
        plt.savefig(save_path, dpi=400)
        plt.show()
        plt.close()
        
    def comp_phases(self):
        
        r_extremes_1 = self.Schr_sol[0].t_events
        r_extremes_2 = self.Schr_sol[1].t_events
        
        def bessel1(r, a, delta):
            k = self.eps_v
            return a * np.sin(k * r - 1/2 * l * np.pi + delta)

        def bessel2(r, a, delta):
            k = np.sqrt(self.eps_v ** 2 - self.eps_delta ** 2)
            return a * np.sin(k * r - 1/2 * l * np.pi + delta)

        if self.update_accuracy:
            try:
                i11 = np.argwhere(self.r >= r_extremes_1[0][-5])[0, 0]  # index in domain where oscillations should start
            except IndexError:
                self.r_max = 2 * self.r_max
                self.solve_Schr_eq()
        else:
            i11 = np.argwhere(self.r >= r_extremes_1[0][-5])[0, 0]
            
        (a11, delta11), cov11 = curve_fit(bessel1, self.r[i11:], self.psi[0][i11:], p0 = (self.psi[0][i11], 0)) 
        delta11 = delta11%(2*np.pi)

        i12 = np.argwhere(self.r >= r_extremes_1[1][-5])[0, 0]
        (a12, delta12), cov12 = curve_fit(bessel2, self.r[i12:], self.psi[1][i12:], p0 = (self.psi[1][i12], 0))
        delta12 = delta12%(2*np.pi)

        i21 = np.argwhere(self.r >= r_extremes_2[0][-5])[0, 0]
        (a21, delta21), cov21 = curve_fit(bessel1, self.r[i21:], self.psi[2][i21:], p0 = (self.psi[2][i21], 0))
        delta21 = delta21%(2*np.pi)

        i22 = np.argwhere(self.r >= r_extremes_2[1][-5])[0, 0]
        (a22, delta22), cov22 = curve_fit(bessel2, self.r[i22:], self.psi[3][i22:], p0 = (self.psi[3][i22], 0))
        delta22 = delta22%(2*np.pi)
        
        self.delta = delta11, delta12, delta21, delta22
        self.beta = a11, a12, a21, a22
        
    def comp_S(self):
        
        F_tilde = np.array([
                    [self.beta[0] * np.exp(-1j * self.delta[0]), self.beta[2] * np.exp(-1j * self.delta[2])],
                    [self.beta[1] * np.exp(-1j * self.delta[1]), self.beta[3] * np.exp(-1j * self.delta[3])]
                    ])
        F = np.dot(np.array([[1, 0], [0, np.sqrt(np.sqrt(self.eps_v**2 - self.eps_delta**2)/self.eps_v)]]), F_tilde)
        self.S = np.dot(np.conjugate(F), np.linalg.inv(F))
        
        self.unitarity = np.linalg.norm(np.eye(2) - np.dot(self.S,self.S.conj().T))
        if self.unitarity > 1e-2 and self.update_accuracy:
            self.num_acc += 1
            self.solve_Schr_eq()
            self.comp_phases()
            self.comp_S()
            
    def comp_cs_Sommerfeld(self):
        
        # The scattering cross-section:
        self.cs = np.pi / np.array([[k_0, k_1], [k_0, k_1]])**2 * (2*l+1) * np.abs(S - np.identity(2))**2
        
        # The Sommerfeld enhancement factor
        self.Sommerfeld = None
        
```
